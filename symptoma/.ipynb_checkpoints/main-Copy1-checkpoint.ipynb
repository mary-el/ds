There are too many different diagnoses, that's because some of them are synonyms; let's read the dictionary of synonyms:

```python
synonyms = json.load(open('data/synonyms.json'))
```

```python
len(synonyms.keys())
```

I'll reverse the dictionary and add the keys as the synonyms for themselves:

```python
synonyms_reversed = {key: name for name, synonyms_list in synonyms.items() for key in synonyms_list}
synonyms_reversed.update({key: key for key in synonyms.keys()})
```

Now I'll turn all the synonyms into just one diagnosis:

```python
checker1_cleaned = checker_1.applymap(lambda x: synonyms_reversed[x])
checker2_cleaned = checker_2.applymap(lambda x: synonyms_reversed[x])
checker3_cleaned = checker_3.applymap(lambda x: synonyms_reversed[x])
```

```python
def get_diag(df):
    diag = df.drop("Actual Diagnosis", axis=1).apply(lambda x: x == df['Actual Diagnosis'])
    return diag
```

```python
diag1 = get_diag(checker1_cleaned)
diag2 = get_diag(checker2_cleaned)
diag3 = get_diag(checker3_cleaned)
```

```python
diag1.head()
```

We suppose that every checker ranked the suggestions by probability: Suggested diagnosis 1 is ranked higher than Suggested Diagnosis 2 etc; also we can see that some of suggestions for one case are the same, so we're intersted only in the first right suggestion. In this case we can use Mean reciprocal rank (MAP) metric.
The mean reciprocal rank is a statistic measure for evaluating any process that produces a list of possible responses to a sample of queries, ordered by probability of correctness. The reciprocal rank of a query response is the multiplicative inverse of the rank of the first correct answer: 1 for first place, 1⁄2 for second place, 1⁄3 for third place and so on. The mean reciprocal rank is the average of the reciprocal ranks of results for a sample of queries Q (https://en.wikipedia.org/wiki/Mean_reciprocal_rank)

```python
def get_MAP(df):
    map_value = 0
    for index, row in df.iterrows():
        map_value +=  (1 / (list(row).index(True) + 1)) if (True in list(row)) else 0
    map_value /= len(df)
    return map_value
```

```python
get_MAP(diag1)
```

Now there are less unique diagnoses.  

We assume that a checker is correct when the right diagnosis is among the checker's suggestions. I'll add a column showing if the checker is correct:

```python
def get_right_diagnosis(df):
    df['right_diagnosis'] = df.drop("Actual Diagnosis", axis=1).isin(df['Actual Diagnosis']).any(1)
```

```python
get_right_diagnosis(checker1_cleaned)
get_right_diagnosis(checker2_cleaned)
get_right_diagnosis(checker3_cleaned)
```

```python
checker1_cleaned['right_diagnosis'].sum()
```

Now we'll create a dataframe of the checkers' results:

```python
right_diagnosis = pd.DataFrame([checker1_cleaned['right_diagnosis'], checker2_cleaned['right_diagnosis'], checker3_cleaned['right_diagnosis']], ['Checker 1', 'Checker 2', 'Checker 3']).T
```

```python
right_diagnosis
```

And calculate the accuracy for each checker:

```python
accuracy_df = right_diagnosis.mean()
accuracy_df
```

We can visualize it with a barplot:

```python
plt.bar(accuracy_df.keys(), accuracy_df.values, color=['r', 'g', 'b'])
```
